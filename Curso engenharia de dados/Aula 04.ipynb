```python
#Fibonacci até o número 100
def fibonacci(c):
    if c==1:
        return 1
    if c==2:
        return 1
    return fibonacci(c-1) + fibonacci(c-2)
```

```python
i=1
while fibonacci(i)<=100:
    print(fibonacci(i))
    i+=1
```

```python
#fatorial
def fat(n):
    if n>1:
        return n*fat(n-1)
    return 1
```

```python
fat(1024)
```
